2) Frecuencia de continent: utilizando value_counts responde lo siguiente:

● ¿Cuál es el continente con una mayor presencia en la muestra?

● ¿Cuál es la probabilidad de elegir un equipo asiático al azar?

● ¿Cuál es la probabilidad de elegir un equipo africano al azar?


```python
serie_continent = df['continent']
serie_continent.value_counts()
```

```python
prob_asia = serie_continent.value_counts('%').get('asia')
print("La probabilidad de que aparezca un equipo asiatico es de : {}%".format(100 * prob_asia))
```

```python
prob_africa = serie_continent.value_counts('%').get('africa')
print("La probabilidad de que aparezca un equipo asiatico es de : {}%".format(100 * prob_africa))
```

4) Probabilidad de clasificación a la siguiente ronda: calcula la probabilidad de que un país pase a la siguiente ronda (clasifique) por continente.

● ¿Cuál es la probabilidad de que un país asiático pase a la siguiente ronda?

● ¿Cuáles son los dos continentes con la mayor probabilidad de clasificar?

● ¿Cuál es la probabilidad de que un país europeo no clasifique?

```python
def calc_prob_clasif(df):
    probabilidades = df['clasificado'].value_counts(normalize = True)
    return probabilidades
```

```python
calc_prob_clasif(asia_df)
```

```python
print('africa')
print(calc_prob_clasif(africa_df))
print(' ')
print('europe')
print(calc_prob_clasif(europe_df))
print(' ')
print('asia')
print(calc_prob_clasif(asia_df))
print(' ')
print('northamerica')
print(calc_prob_clasif(northamerica_df))
print(' ')
print('southamerica')
print(calc_prob_clasif(southamerica_df))
print(' ')
```

```python
continentes_df = [
    africa_df.copy(), 
    europe_df.copy(),
    asia_df.copy(),
    northamerica_df.copy(),
    southamerica_df.copy()
]

for n_iter in continentes_df:
    print(" ")
    print("Nombre Continente:")
    print(n_iter.iloc[0,1])
    print("Probabilidad de Clasificación:")
    print(n_iter['clasificado'].value_counts(normalize=True))
```

```python
probs = df.groupby('continent')['ha_ganado'].value_counts(normalize=True)
```

```python
max_prob = probs.max()
continente_max_prob = probs[probs == max_prob].index[0][0]
```
